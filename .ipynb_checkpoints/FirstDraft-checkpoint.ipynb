```python
#Create new dataframes to calculate differences
serv1 = pd.DataFrame()
serv2 = pd.DataFrame()

#Server 1 columns and values
serv1['price'] = low1['unit_price']
serv1['quantity'] = low1['quantity']
serv1['total'] = low1['unit_price']*low1['quantity']
serv1p = serv1['price'].mean()
serv1 = serv1[ serv1['price'] <= (serv1p) ]
serv1avg = round(serv1['price'].mean(), 2)
serv1quant = sum(serv1['quantity'])
print(f"On server 1, the average cost of the item is {serv1avg:,} and the quantity is {serv1quant:,}.")


# Reset index and drop old index
serv1 = serv1.reset_index()
serv1.drop(['index'], axis=1, inplace=True)

#Server 2 columns and values
serv2['price'] = low2['unit_price']
serv2['quantity'] = low2['quantity']
serv2['total'] = low2['unit_price']*low2['quantity']
serv2p = serv2['price'].mean()
serv2 = serv2[ serv2['price'] <= (serv2p)]
serv2avg = round(serv2['price'].mean(), 2)
serv2quant = sum(serv2['quantity'])
print(f"On server 2, the average cost of the item is {serv2avg:,} and the quantity is {serv2quant:,}.")



# Reset index and drop old index
serv2 = serv2.reset_index()
serv2.drop(['index'], axis=1, inplace=True)
```

```python
#Calculate the total cost of buying out server 1 and 2
buyout1 = round(sum(serv1['total'])/10000, 3)
buyout2 = round(sum(serv2['total'])/10000, 3)

#Calculate the quantity * average cost for both scenarios
conversion1 = serv1quant * serv2avg
conversion2 = serv2quant * serv1avg

#Factor in the cost of buyout with the potential earning
pe1 = round(conversion1 - buyout1, 3)
pe2 = round(conversion2 - buyout2, 3)

#Factor in the cost of 2 WoW tokens
tcost1 = round(pe1 - (token_gold*2),3)
tcost2 = round(pe2 - (token_gold*2),3)
```

```python
print(f"To buyout server 1, it would cost: {buyout1:,}")
print(f"To buyout server 2, it would cost: {buyout2:,}")
print("This is how much profit you could make moving from server 1 to 2:")
print(f"{pe1:,} gold")
print("Factoring 2 WoW tokens for transfer, you would net:")
print(f"{tcost1:,} gold")
print("This is how much profit you could make moving from server 2 to 1:")
print(f"{pe2:,} gold")
print("Factoring 2 WoW tokens for transfer, you would net:")
print(f"{tcost2:,} gold")
```
